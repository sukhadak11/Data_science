```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sales_data_with_discounts.csv")
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
# separt by numerical and categarical columes
#Identify Data Types
numerical_cols = df.select_dtypes(include=['int64','float64']).columns
categarical_cols = df.select_dtypes(include=['object']).columns
```

```python
print("Numerical Columns",list(numerical_cols))
print("Categarical Colums",list(categarical_cols))
```

```python
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols]).mean()
```

```python
for col in categarical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
print("After using data prepossing")
df.isnull().sum()
```

```python
print("number of duplicate row", df.duplicated().sum())
```

```python
print("remove duplicate", df.drop_duplicates())
```

This code:


Loads and displays the dataset.

Checks for missing values and data types.

Cleans the dataset (fills missing data, removes duplicates).

Prepares the data for descriptive analytics and visualization (next steps).

Descriptive Analytics for Numerical Columns

Step 1: Calculate Mean, Median, Mode, and Standard Deviation

```python
for col in numerical_cols:
    mean_val = df[col].mean()
    median_val = df[col].median() 
    mode_val = df[col].mode()[0]
    std_val = df[col].std()
    print(f"{col}:")
    print(f"  Mean = {mean_val:.2f}")
    print(f"  Median = {median_val:.2f}")
    print(f"  Mode = {mode_val}")
    print(f"  Standard Deviation = {std_val:.2f}\n")
```

# Data Visualization

```python
df[numerical_cols].hist(figsize=(12, 8), bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Histogram for Numerical Columns", fontsize=14)
plt.show()
```

# Box Plots (Outlier Detection)

Create boxplots for numerical variables to identify outliers and the interquartile range.
Discuss any findings, such as extreme values or unusual distributions.
    

```python
plt.figure(figsize=(12,6))
for i,col in enumerate(numerical_cols,1):
    plt.subplot(1,len(numerical_cols), i)
    sns.boxplot(y=df[col], color='green')
    plt.title(f"{col}   Boxplot")
plt.tight_layout()    
plt.show()    
```

Boxplots show the spread (IQR, quartiles, outliers) of data.

In dataset:

All values (e.g., Volume, Avg Price, Total Sales Value, Discount Rate, etc.) seem to be almost constant.

There is no visible variance or outliers, so the boxplot becomes a straight line.

# Bar Chart

Bar Chart Analysis for Categorical Column: Identify categorical columns in the dataset. Create bar charts to visualize the frequency or count of each category. Analyze the distribution of categories and provide insights.

```python
categarical_cols = df.select_dtypes(include=['object']).columns
print("Categarical Columns:", list(categarical_cols))
```

```python
for col in categarical_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(x=col, data=df, palette='pastel',edgecolor='black')
    plt.title(f"Categorical Column {col}")
    plt.xticks(rotation=45)
    plt.show()    
```

Conclusion:

- In the histograms for all numerical columns (such as Volume, Avg Price, Total Sales Value, Discount Rate, Discount Amount, and Net Sales Value) show only one single bar instead of a spread of values. This means that all values in each numerical column are almost the same, with very little or no variation across records.
