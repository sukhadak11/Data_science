# Discuss the scenarios where each scaling technique is preferred and why.

Ans: 
- Min-Max Scaling is preferred when the model is sensitive to the magnitude of features or requires values within a fixed 0–1 range, such as KNN, SVM, Neural Networks, and other distance-based algorithms. It is also useful when the data does not follow a normal distribution. 
- In contrast, Standard Scaling is preferred when the data is roughly normally distributed or when algorithms rely on variance and assume standardized input, such as Logistic Regression, Linear Regression, PCA, and K-Means. It is also more suitable when outliers are present, as it does not compress values into a limited range.

# Discuss the pros and cons of One-Hot Encoding and Label Encoding.

Ans:
- One-Hot Encoding avoids ordinal relationships and is suitable for most ML models but increases dimensionality when categories are many. 
- Label Encoding is simple and memory-efficient but introduces artificial ordering between categories, which can mislead models that assume numerical meaning. Tree-based models generally handle label encoding well, while linear and distance-based models work better with One-Hot Encoding.

## Apply a transformation (e.g., log transformation) to at least one skewed numerical feature and justify your choice.

```python
import numpy as np
# Apply log transformation to the skewed feature
# log(1 + x)
df['log_capital_gain'] = np.log1p(df['capital_gain']) 
df[['capital_gain', 'log_capital_gain']]
```

The capital_gain feature is highly skewed, with most individuals having zero gain and a few having extremely high values. To reduce this skewness and stabilize variance, a log(1 + x) transformation was applied. This transformation makes the distribution more normal, reduces the impact of outliers, and helps machine learning models learn patterns more effectively.

```python
import matplotlib.pyplot as plt

# Original and transformed values
orig = df['capital_gain']
log_transformed = np.log1p(orig)

# Plot 1: Original distribution
plt.figure(figsize=(8,5))
plt.hist(orig, bins=50)
plt.title("Original Distribution of capital_gain")
plt.xlabel("capital_gain")
plt.ylabel("Frequency")
plt.show()
```

Original Distribution (Highly Skewed)

- Most values are 0

- A few values go up to 100,000+

- Distribution is extremely right-skewed

- This can hurt model performance

```python
# Plot 2: Log-transformed distribution
plt.figure(figsize=(8,5))
plt.hist(log_transformed, bins=50)
plt.title("Log-Transformed Distribution of capital_gain")
plt.xlabel("log(1 + capital_gain)")
plt.ylabel("Frequency")
plt.show()
```

Log-Transformed Distribution (Reduced Skewness)

- Applying log(1 + capital_gain) compresses large values

- The scale becomes more balanced

- Makes the feature more useful for models like Logistic Regression, SVM, and Neural Networks

- Helps stabilize variance and reduces outlier influence
