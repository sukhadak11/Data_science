```python
import numpy as np
import pandas as pd
```

Step 2: Loading Dataset

Loads your advertising dataset, which likely contains:

User details (Age, Gender, Income, Internet Usage)

Whether they clicked on an ad (Clicked on Ad = 1 or 0)

```python
df = pd.read_csv('/content/Data science/data/Advertising_CL.csv')
df
```

Step 3:
Concept:

Hour: extracts the hour (e.g., 14 means 2 PM)

Day: Monday=0 … Sunday=6

✅ Why?

People behave differently depending on the time and day.

Ex: Users click more ads during evening leisure hours.

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='%d-%m-%Y %H:%M')
df['Hour'] = df['Timestamp'].dt.hour
df['Day'] = df['Timestamp'].dt.dayofweek
```

Step 4:

Features (X): variables used to make predictions.

Target (y): what you’re trying to predict (1 = clicked, 0 = not clicked).

```python
X = df[['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage', 'Male', 'Hour', 'Day']]
y = df['Clicked on Ad']
```

Step 5: Train–Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Step 6: Logistic Regression Model

Concept:
Logistic Regression predicts a binary outcome (0 or 1) using a sigmoid function:

p(y) = 1/1+e^(b0+b1x1)

It outputs a probability between 0 and 1.

0.5 → Clicked (1)

<0.5 → Not clicked (0)

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)
```

Step 7: Model Prediction & Accuracy

```python
y_train_pred = log_reg.predict(X_train)
y_test_pred = log_reg.predict(X_test)
```

It measures how many predictions were correct.


Example:

If 80/100 test samples are correctly classified → Accuracy = 80%.

```python
from sklearn.metrics import accuracy_score
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test,y_test_pred)
```

Step 8: ShuffleSplit Cross Validation

Concept:
It randomly splits your dataset 10 times into train–test sets.
Each time → model is trained and tested → results averaged.

Purpose: Get more reliable accuracy by testing model stability across multiple random splits.

Real-life analogy:
You don’t judge a student by one exam — test multiple times to ensure consistent performance.

```python
from sklearn.model_selection import ShuffleSplit,cross_validate
ShuffleSplit = ShuffleSplit(n_splits=10, test_size=0.3, random_state=42)
```

Step 9: K-Fold Cross Validation

Concept:

The dataset is divided into K parts (folds) → each part becomes test set once, remaining K−1 parts are training sets.

Purpose: Reduces bias due to one lucky split.

Example:
If K=5, every data point gets to be in the test set exactly once.

```python
from sklearn.model_selection import KFold,cross_validate
KFold = KFold(n_splits=5, shuffle=True, random_state=42)
```

Step 10: Comparing the Results

Train accuracy → how well model learns.

Test accuracy → how well model generalizes.

```python
# You need to run cross_validate first to get the scores
# Example using KFold:
cv_results = cross_validate(log_reg, X, y, cv=KFold, scoring='accuracy', return_train_score=True)

print('cross validation: train accuracy', round(cv_results['train_score'].mean(), 2))
print('cross validation: test accuracy', round(cv_results['test_score'].mean(), 2))
```

| Concept             | Explanation             | Real-life Analogy      |
| ------------------- | ----------------------- | ---------------------- |
| Train–Test Split    | Divides data once       | One exam               |
| ShuffleSplit        | Repeats random splits   | Multiple random exams  |
| K-Fold              | Systematic splits       | Everyone gets a chance |
| Logistic Regression | Predicts binary outcome | Yes/No decision        |
| Cross Validation    | Checks consistency      | Reliability check      |
