```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import scatter_matrix
```

```python
df = pd.read_csv("Cardiotocographic.csv")
df
```

```python
df.head()
```

```python
print("Data information", df.info())
print("\nData shape", df.shape)
print("\nMissing Values:\n", df.isnull().sum())
print("\nColumns name", df.columns)
```

```python
print("\nData Describetion", df.describe())
```

```python
# Example: try to convert every object column to numeric when possible
for col in df.columns:
    if df[col].dtype == 'object':
        # strip whitespace then try convert
        temp = df[col].str.strip()
        coerced = pd.to_numeric(temp, errors='coerce')
        convert_ratio = coerced.notna().mean()   # fraction converted
        if convert_ratio > 0.6:                 # heuristic threshold
            df[col] = coerced
            print(f"Converted {col} to numeric ({convert_ratio:.2%} values converted)")
```

```python
df.info()
```

```python
# Fill missing numeric values using median
df = df.fillna(df.median())
df
```

```python
df.isnull().sum()
```

```python
#Outlier & cap Outliers using IQR
for col in df.columns:
    Q1 = df[col].quantile(0.5)
    Q3 = df[col].quantile(0.75)
    IQR= Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower,upper)
df.columns
```

1. MSTV and MLTV move together

- Both show how the fetal heart rate changes over time.

- When short-term variability (MSTV) increases, long-term variability (MLTV) also increases.

- This means the baby’s heart activity is consistent across different time scales.

2. ASTV and ALTV are related

- ASTV = abnormal short-term variability

- ALTV = abnormal long-term variability

- If the baby shows abnormal short-term behavior, they often show abnormal long-term behavior too.

3. Width is linked with variability features

- Width becomes higher when variability (ASTV, MLTV) increases.

- A wider heart rate range usually means the baby is more active or stressed.

4. Deceleration features (DL, DS, DP) are connected

- If one type of deceleration happens, others may appear too.

- These features tend to occur together.

- This may indicate fetal stress.

5. NSP (the fetal health label) depends on variability

- NSP is related to ASTV, ALTV, and Width.

- Higher abnormal variability often leads to NSP = “suspect” or “pathologic.”

- Variability features are important for predicting fetal health.

6. LB (baseline heart rate) does not relate strongly with others

- LB stays steady and does not change much with other features.

- Baseline heart rate alone cannot tell us the baby's health condition.

B. Time-Based Patterns

- This dataset does not contain a time column (like seconds, minutes, or dates).

- Each row is a single measurement, not part of a time sequence.

- So, we cannot study trends over time.
- We can only compare measurements across different patients.
