# 3. Hyperparameter Tuning
   
●	Modify various hyperparameters, such as the number of hidden layers, neurons per hidden layer, activation functions, and learning rate, to observe their impact on model performance.

●	Adopt a structured approach like grid search or random search for hyperparameter tuning, documenting your methodology thoroughly.


```python
def build_model(neurons=32, activation='relu', lr=0.001):
    model = Sequential()
    model.add(Dense(neurons, activation=activation,
                    input_shape=(X_train.shape[1],)))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(
        optimizer=Adam(learning_rate=lr),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model
```

```python
neurons_list = [16, 32, 64]
activations = ['relu', 'tanh']
learning_rates = [0.01, 0.001]

results = []
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
for n in neurons_list:
    for act in activations:
        for lr in learning_rates:
            model = build_model(neurons=n, activation=act, lr=lr)

            history = model.fit(
                X_train, y_train,
                epochs=20,
                batch_size=16,
                validation_split=0.2,
                verbose=0
            )

            val_acc = max(history.history['val_accuracy'])
            results.append((n, act, lr, val_acc))
```

```python
best = max(results, key=lambda x: x[3])
print("Best Neurons:", best[0])
print("Best Activation:", best[1])
print("Best Learning Rate:", best[2])
print("Best Validation Accuracy:", best[3])
```
